```python
import pandas as pd
df=pd.read_csv("/content/auto-mpg.csv")
```

```python
print("first 10 rows:")
print(df.head(10))
print("\nShape of dataset(rows,colomns):")
print(df.shape)
print('\nSummary Statistics:')
print(df.describe())
```

```python
filtered = df[(df['model year'] == 75) & (df['weight'] < 3000)]
result = filtered[['car name', 'weight', 'mpg']]
print(result)
```

```python
print(df.isnull().sum())
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
print(df[['horsepower','weight','power_to_weight_ratio']].head())
```

```python
grouped = df.groupby('origin')['mpg'].mean()
print(grouped)
```

```python
top10 = df.sort_values(by='mpg', ascending=False).head(10)
print(top10[['car name', 'mpg']])
```

```python
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']
df['performance_score'] = df.apply(performance_score, axis=1)
print(df[['car name', 'mpg', 'acceleration', 'weight', 'performance_score']].head())
```

```python
summary = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
print(summary)
```

```python
import pandas as pd
df = pd.read_csv("auto-mpg.csv")
df.columns = df.columns.str.strip().str.replace(" ", "_")
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv("high_mpg_cars.csv", index=False)
print("File saved successfully ✅")
print(high_mpg.head())
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("auto-mpg.csv")
acc = df['acceleration'].values
normalized_acc = (acc - acc.min()) / (acc.max() - acc.min())
print("Normalized values:")
print(normalized_acc[:5])
```

```python
df = pd.read_csv("auto-mpg.csv")
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
hp = df['horsepower'].values
hp_mean = np.nanmean(hp)
hp = np.where(np.isnan(hp), hp_mean, hp)
hp_new = hp * 1.10
print("Updated horsepower values (first 5):")
print(hp_new[:5])
```

```python
displacement = df['displacement'].values
origin = df['origin'].values
avg_disp = np.mean(displacement[origin == 2])
print(avg_disp)
```

```python
df = pd.read_csv("auto-mpg.csv")
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mean())
matrix = df[['mpg', 'horsepower', 'weight']].values
vector = np.array([1, 0.5, -0.2])
result = np.dot(matrix, vector)
print("Dot product result (first 5 values):")
print(result[:5])
```

```python
mpg = df['mpg'].values
weight = df['weight'].values
corr = np.corrcoef(mpg, weight)[0,1]
print(corr)
```
